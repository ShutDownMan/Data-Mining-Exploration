```python
import dask.dataframe as dd

data_size = "full"
root_path = f"./"
data_path = f"{root_path}/data/{data_size}/"
```

```python
ddf = dd.read_csv(f"{data_path}/credit_dataset.csv")

ddf = ddf.drop(columns=["Unnamed: 0", "ID"])

ddf.head()
```

```python
ddf["GENDER"] = ddf["GENDER"].map(lambda x: 0 if x == "M" else 1)
ddf["CAR"] = ddf["CAR"].map(lambda x: 0 if x == "N" else 1)
ddf["REALITY"] = ddf["REALITY"].map(lambda x: 0 if x == "N" else 1)

ddf.head()
```

```python

int_inc_type = dd.get_dummies(ddf["INCOME_TYPE"].compute(), prefix="INC_TYPE")
int_edu_type = dd.get_dummies(ddf["EDUCATION_TYPE"].compute(), prefix="EDU_TYPE")
int_fam_type = dd.get_dummies(ddf["FAMILY_TYPE"].compute(), prefix="FAM_TYPE")
int_hou_type = dd.get_dummies(ddf["HOUSE_TYPE"].compute(), prefix="HOU_TYPE")

int_inc_type.head()
```

```python
col_remove = ["INCOME_TYPE", "EDUCATION_TYPE", "FAMILY_TYPE", "HOUSE_TYPE"]
ddf = ddf.drop(columns=col_remove)

ddf.head()
```

```python
ddf_full = dd.concat([ddf, int_inc_type, int_edu_type, int_fam_type, int_hou_type], axis=1)

ddf_full.head()
```
